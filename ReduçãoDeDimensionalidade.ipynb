```python
import matplotlib.pyplot as plt
import numpy as np

img = plt.imread('/content/Cosmic Serpent Print.jfif')

plt.imshow(img)
plt.axis('off')  # Remove os eixos
plt.show()

# Separao dos canais RGB
vermelho = img[:, :, 0]
verde = img[:, :, 1]
azul = img[:, :, 2]

img_cinza = 0.2126 * vermelho + 0.7152 * verde + 0.0722 * azul

plt.imshow(img_cinza, cmap='gray')  # Define o mapa de cores como cinza
plt.axis('off')
plt.show()

limiar = 108
img_binaria = np.zeros_like(img_cinza)  # Cria uma matriz 2D com as mesmas dimenses de img_cinza

for linha in range(img_cinza.shape[0]):
    for coluna in range(img_cinza.shape[1]):
        if img_cinza[linha, coluna] < limiar:
            img_binaria[linha, coluna] = 0
        else:
            img_binaria[linha, coluna] = 1

plt.imshow(img_binaria, cmap='gray')
plt.axis('off')
plt.show()
```
